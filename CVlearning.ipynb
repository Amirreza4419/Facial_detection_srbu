Show it in matplotlib :

```python
plt.imshow(c_img)
```

the color are BGR instead of RGB in open cv so we reverse it :

```python
c_img_r = c_img[: , : , ::-1]
plt.imshow(c_img_r)
```

```python
#split
c_img_bgr = c_img
b , g ,r = cv.split(c_img_bgr)
#show
plt.figure(figsize=[20,5])

plt.subplot(141);plt.imshow(r , cmap='gray');plt.title("red")
plt.subplot(142);plt.imshow(g , cmap='gray');plt.title("green")
plt.subplot(143);plt.imshow(b , cmap='gray');plt.title("blue")

#megre
c_img_merge = cv.merge((b,g,r))
#show
plt.subplot(144)
plt.imshow(c_img_merge[:,:,::-1])
plt.title("merged")
```
